```python
df.set_index(df['отправление'], inplace=True)
```

```python
code = pd.read_excel('Копия ЖД СТАНЦИИ СПРАВОЧНИК 6.xls',dtype=str, sheet_name=0)
```

```python
code.set_index('название', inplace=True)
```

```python
staition_begin =  df.join(code)
staition_begin.rename(columns={'код':'код станции отправления'}, inplace= True)
```

```python
staition_all =  staition_begin.join(code)
staition_all.rename(columns={'код':'код станции назначения'}, inplace= True)
```
